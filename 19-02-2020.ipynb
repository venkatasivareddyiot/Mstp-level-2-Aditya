```python
mycontacts={}
```

```python
#Add Contact
def addContact(name,no):
    if name not in mycontacts:
        mycontacts[name]=no
        print("contact added")
    else:
        print("contact already exists")
```

```python
addContact("",12345645878)
```

```python
mycontacts
```

```python
def searchContact(name):
    if name in mycontacts:
        print(name,":",mycontacts[name])
    else:
        print("contact not found")
```

```python
searchContact("teju")
```

```python
#edit Contact
def editContact(name,no):
    if name in mycontacts:
        mycontacts[name]=no
        print("contact edited")
    else:
        print("contact not found")
        
```

```python
editContact("teju",9581645635)
```

```python
mycontacts
```

```python
#delete a contact
def deleteContact(name):
    if name in mycontacts:
        mycontacts.pop(name)
        print("contact deleted")
    else:
        print("contact not found")      
```

```python
mycontacts
```

```python
deleteContact("siri")
```

```python
def phoneValidator(no):
    sn=str(no)
    l=len(sn)
    if (l==10 and(sn[0]=='9' or sn[0]=='8'or sn[0]=='7' or sn[0]=='6')):
        return True
    if(l==11 and sn[0]=='0' and (sn[1]=='9' or sn[1]=='8'or sn[1]=='7' or sn[1]=='6')):
        return True
    if(l==13 and sn[0]=='+' and sn[1:3] =='91' and (sn[3]=='9' or sn[3]=='8'or sn[3]=='7' or sn[3]=='6')):
       return True
    else:
       return False
```
