```python
set1 = {1,2,3,4}
set2 = {4,5,6,7}
set3 = set1 | set2
set3
```

```python
set3.pop()
```

```python
set4 = set1 & set2
set4
```

```python
a = {10,11,12,20,21}
b = {11,13,14,10}
a - b
a.difference(b)
```
