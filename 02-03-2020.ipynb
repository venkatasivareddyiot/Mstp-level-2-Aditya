```python
keypath='DataFiles/key.txt'
```

```python
def keydictionary(path):
    key={}
    with open(path,'r') as f:
        for line in f:
            line=line.split()
            key[line[0]]=line[1]
    return key
```

```python
key=keydictionary(keypath)
```

```python
encryptedpath='DataFiles/encrypteddata.txt'
```

```python
def dataencryption(keypath,datapath):
    key=keydictionary(keypath)
    with open(datapath,'r') as f:
        filedata=f.read().split()
    with open("DataFiles/encrypteddata.txt","w") as f:
        for i in filedata:
            s=''
            for s in i:
                s+=key[s]
            f.write(s+'\n')
    
```

```python
dataencryption(keypath,datapath)
```

```python
def datadecryption(keypath,encryptedpath):
    key=keydictionary(keypath)
    newkey={}
    for key,value in key.items():
        newkey[value]=key
    with open(encryptedpath,'r') as f:
        endata=f.read().split()
    with open("DataFiles/decrypteddata.txt",'w') as f:
        for i in endata:
            f.write(newkey[i]+'\n')
 
```
